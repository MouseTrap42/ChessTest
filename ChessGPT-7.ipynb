```python
# @hidden_cell

import openai
import os
import getpass
import chess
import time

from IPython.core.display_functions import display
from IPython.display import clear_output

#instructions to open project #first download all code as zip, unzip file should now be under file directory in downloads or wherever you saved it to 
#then download latest python from python.org (need it to run python)

#open terminal, type "pip3 install Jupyter Notebook" #after installing, type "Jupyter Notebook" to open Jupyter Notebook 
#Jupyter Notebook should open in web browser, from their find project folder and file in your file directory (ends in .ipynb) and open 
#click run to play chess vs GPT3!

#note: you need your own openai api key (can get from openai, sign in for free with google and click profile to access secret api key), 
#copy and paste key as a string delimited by " " to replace YOUR_API_KEY in code before running

#from tutorial by https://towardsdatascience.com/gpt-3-play-chess-d123a96096a9



#replace YOUR_API_KEY with your own API Key from openai, sign in and retrieve secret key from your profile
openai.api_key = YOUR_API_KEY


completion = openai.Completion.create(
    engine="davinci",
    n=5,
    max_tokens=6,
    prompt="Play chess 1.e4"
)

def get_next_move(prompt):

  #giving the prompt to get the next step. increase n if you want GPT-3 to return more options
  completion = openai.Completion.create(
  engine="text-davinci-002",
  n=50,max_tokens=6,stop=["\n"],prompt= prompt,temperature=1,top_p=1,presence_penalty=0,frequency_penalty=0,echo=True,
)
  choices = completion["choices"]
  moves = []

  #Extract the next step from GPT-3 output
  for choice in choices:
    move = choice["text"].replace(prompt,"")
    if move.strip():
      moves.append(move.split()[0])

  return moves

count = 0
prompt =  "Play chess "
checkmate = False
board = chess.Board()
display(board)

print("moves should be entered as (Pos1Pos2) ie (e2e4)")
while(True):

  while(True):
    move = input("Enter your move  ")
    try:

      # make move
      board.push_san(move)
      clear_output(wait=True)
      display(board)

      # Wait one second before getting GPT-3 move
      time.sleep(1)

      # check for checkmate
      if(board.is_checkmate()):
        print("Checkmate!")
        checkmate = True
        break
      count += 1
      break
    except:
      # if move is illegal prompt again
      print("Illegal move. Try again")

  if(checkmate):
    break
  prompt += str(count) + "." + move
  attempts = 0

  print(prompt)

  # get GPT3 move
  moves = get_next_move(prompt)
  legal_move = False

  for move in moves:
    print(move)

    try:

      # make move
      board.push_san(move)
      clear_output(wait=True)
      display(board)

      # check for checkmate
      if(board.is_checkmate()):
        print("Checkmate!")
        checkmate = True
        break

      legal_move = True
      prompt+= " " + move + "\n"
      break

    except:
      continue

  if(checkmate):
    break

  if not legal_move:
    print("No legal moves returned by GPT-3")
    break



# See PyCharm help at https://www.jetbrains.com/help/pycharm/
```
